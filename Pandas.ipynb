## Crie uma fun√ß√£o lambda que converta a temperatura de Celsius para Fahrenheit

```python
bikes = df.copy()
```

```python
bikes['tempF'] = bikes.apply(lambda t: (t['temp'] * 9/5) + 32, axis=1) # Tem que usar o axis
bikes.head()
```

```python
# Usando assign (Essa maneira me obriga a usar nomes de coluna sem espa√ßos?)
bikes.assign(temp_fahrenheit = lambda c: (c.temp * 9/5) + 32)
```

*Por algum motivo a contagem de elementos durante o slicing no loc √© diferente da forma padr√£o do Python e do iloc, j√° que ele considera o √∫ltimo n√∫mero dado no range;<br>
